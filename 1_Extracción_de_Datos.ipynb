```python
# Convertir en dataframe
df=pd.read_csv("CountryTable.csv")
# df
```

**FILTROS BASICOS ARA VARIABLES CUALITATIVAS**

```python
# El filtro 1 "filtro por objeto" se aplico en la columna "continent" para obtener los registros que solo contengan la categoria "Africa"
filtro1=df[df["continent"]=="Africa"]
filtro1
```

```python
# El filtro 2 "filtro por filas" se aplico en las filas 5 a 23 no consecutivamente
# filtro2=df.iloc[300:302,:]
filtro2=df.iloc[ 5:23 , : ] #filas no consecutivas
filtro2
```

```python
# El filtro 3 "filtro por columnas" se aplico en las columnas 3, 6 y 9 no consecutivamente
# filtro3=df.iloc[:,4:7]
filtro3=df.iloc[:,[3,6,9]] #columnas no consecutivas
filtro3
```

```python
df= pd.read_csv("CountryTable.csv", index_col=2)
df
```

```python
# El filtro 4 "filtro de filas vs columnas" se aplico en la columna "gnp" para obtener los registros que solo contengan la categoria "Africa" o "Asia"

# Paso 2: Se aplico el filtro de filas vs columnas
filtro4=df.loc[["Africa","Asia"], ["gnp"]]
filtro4
```

```python
# El filtro 5 "filtro por cabecera" se aplico en la cabecera
filtro5=df.head(10)
filtro5
```

**FILTROS BASICOS PARA VARIABLES CUANTITATIVAS**

```python
# El filtro 6 "filtro por comparacion" se aplico en la columna "gnp" para obtener los registros que solo contengan datos menor a "2005"
filtro6=df[df["gnp"]<2005]
filtro6
```

```python
# El filtro 7 "filtro Y" se aplico en la columna "gnp" para obtener los registros mayores a "200000" y "name" igual a "Belgium"
filtro7=df[(df["population"]>200000) & (df["name"]=="Belgium")]
filtro7
```

```python
# El filtro 8 "filtro O" se aplico en la columna "gnp" para obtener los registros que solo contengan la categoria mayor a "200000" o "name" igual a "Belgium"
filtro8=df[(df["population"]>200000) | (df["name"]=="Belgium")]
filtro8
```

```python
# El filtro 9 "filtro NOT" se aplico en la columna "gnp" para no obtener los registros que contengan la categoria mayor o igual "200000" y tampoco "name" igual a "Belgium"
filtro9=df[~(df["population"]>=200000) & ~(df["name"]=="Belgium")]
filtro9
```

```python
#Convertir archivo filtrado a CSV
filtro9.to_csv("Entregable1.csv")
```
